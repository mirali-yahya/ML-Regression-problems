### Outlier Treatment
#### With the help of IQR (Interquartile Range) method, we will "squeeze" the outlier values.

```python
# Function to find upper and lower bound of the dataset
def outlier_detect(cols):
    Q1, Q3 = np.percentile(cols,[25, 75])
    IQR = Q3 - Q1
    upper_bound = Q3 + (1.5 * IQR)
    lower_bound = Q1 - (1.5 * IQR)
    return upper_bound, lower_bound
```

```python
# Seperate Numerical and Categorical variables. We will need Numeric variables to treat outliers
num_vars = []
cat_vars = []

for i in df.columns:
    if type(i) == str:
        cat_vars.append(i)
    else:
        num_vars.append(i)
```

```python
# Checking and removing outliers
for i in num_vars:
    upper, lower = outlier_detect(df[i])
    df[i] = np.clip(df[i], a_min = lower, a_max = upper)
```
